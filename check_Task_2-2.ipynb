```python
# Đường dẫn tới file Parquet
file_path = './results/Task_2-2.parquet'

# Đọc file bằng pyarrow hoặc fastparquet
df = pd.read_parquet(file_path)

# Hiển thị 10 dòng đầu tiên
df.head(10)
```
