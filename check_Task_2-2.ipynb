```python
import pandas as pd
from IPython.display import display

# 1. Đọc và chuẩn hóa toàn bộ file
df_mine = pd.read_parquet('results/Task_2-2.parquet')
df_tp = pd.read_parquet('results/Task_2-2-TP.parquet')

df_mine.columns = [c.lower() for c in df_mine.columns]
df_tp.columns = [c.lower() for c in df_tp.columns]

if pd.api.types.is_numeric_dtype(df_mine['month']):
    df_mine['month'] = '2022-' + df_mine['month'].astype(str).str.zfill(2)

df_mine['sku'] = df_mine['sku'].astype(str)
df_mine['month'] = df_mine['month'].astype(str)
df_tp['sku'] = df_tp['sku'].astype(str)
df_tp['month'] = df_tp['month'].astype(str)

# 2. HÀM TRA CỨU CHI TIẾT
def xem_chi_tiet(target_sku, target_month):
    """
    Hiển thị dữ liệu của một cặp (SKU, Month) cụ thể trên cả 2 file
    """
    print(f"{'='*60}")
    print(f" ĐANG TRA CỨU: SKU = '{target_sku}', Month = '{target_month}'")
    print(f"{'='*60}")
    
    # Lọc dữ liệu
    row_mine = df_mine[(df_mine['sku'] == target_sku) & (df_mine['month'] == target_month)]
    row_tp = df_tp[(df_tp['sku'] == target_sku) & (df_tp['month'] == target_month)]
    
    if row_mine.empty and row_tp.empty:
        print("❌ Không tìm thấy dữ liệu ở cả 2 file.")
        return
        
    print("\n👉 [PHIÊN BẢN 1] - FILE CỦA BẠN (Task_2-2.parquet)")
    if not row_mine.empty:
        display(row_mine)
    else:
        print("   ❌ (Không tồn tại trong file của bạn)")
        
    print("\n👉 [PHIÊN BẢN 2] - FILE MẪU TP (Task_2-2-TP.parquet)")
    if not row_tp.empty:
        display(row_tp)
    else:
        print("   ❌ (Không tồn tại trong file mẫu)")

# 3. CHẠY THỬ HÀM (Bạn có thể đổi SKU và Month tùy ý)
# Ví dụ: lấy ra 1 SKU đang bị lệch kết quả để "bắt bệnh"
xem_chi_tiet(target_sku="SET264-KR-NP-XL", target_month="2022-04")
```
